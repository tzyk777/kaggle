```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns = 100
```

```python
train_data=pd.read_csv('train_ver2.csv')
```

```python
train_data.shape
```

```python
production=train_data.ix[:,24:48]
```

```python
counts=production.apply(pd.Series.value_counts)
counts
```

```python
counts.transpose().plot(kind='bar',stacked=True, legend=False)
plt.legend(('Not Bought','Bought'),loc='best')
plt.title('Production Bought Distribution')
plt.xlabel('production')
plt.ylabel('number of people')
plt.show()
```

"Current Account" is the most popular production.
"Guarantee" is the least popular production.

```python
correlation=production.corr()
correlation
```

```python
heatmap=plt.pcolor(correlation)
plt.colorbar(heatmap)
plt.show()
```

High Correlations:
corr(Payroll Account, Payroll)=0.76
corr(Payroll Account, Pensions)=0.79
corr(Payroll, Pensions)=0.96
